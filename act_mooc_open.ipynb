```python
import pandas as pd

features_dir = 'datasets/act-mooc/mooc_action_features.tsv'
labels_dir = 'datasets/act-mooc/mooc_action_labels.tsv'
actions_dir = 'datasets/act-mooc/mooc_actions.tsv'

features_df = pd.read_csv(features_dir, sep='\t')
labels_df = pd.read_csv(labels_dir, sep='\t')
actions_df = pd.read_csv(actions_dir, sep='\t')
```

```python
features_df["labels"] = labels_df["LABEL"]
features_df
```

```python
features_df["user_id"] = actions_df["USERID"]
features_df["target_id"] = actions_df["TARGETID"]
features_df["time"] = actions_df["TIMESTAMP"]
features_df = features_df.drop(columns=["ACTIONID"])
features_df
```

```python
column_order = ["user_id", "target_id", "time", "labels", "FEATURE0", "FEATURE1", "FEATURE2", "FEATURE3"]
features_df = features_df[column_order]
features_df
```

```python
pd.set_option('display.max_rows', 20)

features_df["time_delta"] = 0
```

```python
from tqdm import tqdm

row_last_time = [0 for _ in range(7047)]

for index, row in tqdm(features_df.iterrows()):
    user_id = int(row["user_id"])
    time = row["time"]

    features_df.at[index, "time_delta"] = time - row_last_time[user_id]
    row_last_time[user_id] = time
```
