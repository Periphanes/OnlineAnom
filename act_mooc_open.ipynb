```python
import pandas as pd

features_dir = 'datasets/act-mooc/mooc_action_features.tsv'
labels_dir = 'datasets/act-mooc/mooc_action_labels.tsv'
actions_dir = 'datasets/act-mooc/mooc_actions.tsv'

features_df = pd.read_csv(features_dir, sep='\t')
labels_df = pd.read_csv(labels_dir, sep='\t')
actions_df = pd.read_csv(actions_dir, sep='\t')
```

```python
features_df["labels"] = labels_df["LABEL"]
features_df
```

```python
features_df["user_id"] = actions_df["USERID"]
features_df["target_id"] = actions_df["TARGETID"]
features_df = features_df.drop(columns=["ACTIONID"])
features_df
```

```python
column_order = ["user_id", "target_id", "labels", "FEATURE0", "FEATURE1", "FEATURE2", "FEATURE3"]
features_df = features_df[column_order]
features_df
```

```python
import numpy as np
data_arr = features_df.to_numpy()
data_arr
```
